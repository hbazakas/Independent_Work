```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
games = pd.read_csv("games.csv")
games = games.dropna()
games["PTS_total"] = games["PTS_home"] + games["PTS_away"]
games["AST_total"] = games["AST_home"] + games["AST_away"]
games["REB_total"] = games["REB_home"] + games["REB_away"]
games = games[['HOME_TEAM_ID','VISITOR_TEAM_ID', 'SEASON', 'PTS_home', 'FG_PCT_home',
       'FT_PCT_home', 'FG3_PCT_home', 'AST_home', 'REB_home', 'TEAM_ID_away',
       'PTS_away', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away', 'AST_away',
       'REB_away', 'PTS_total', 'AST_total', 'REB_total', 'HOME_TEAM_WINS']]

seasons = np.sort(games.SEASON.unique())
```

```python

def corr_year(year, df):
    index = (year-2002)*19
    
    print(df.groupby("SEASON").corr().HOME_TEAM_WINS[index-19:index])


```

```python
for i in seasons:
    corr_year(i, games)
```

* Rebounds seem to be the next most correlated, followed by AST.
* PCT stats show a consistent correlation with winning, albeit smaller in magnitude than the aforementioned categories. However, this is of course skewed by scale, as PTS can change by a lot while PCT stats are bounded on [0,1] Normalizing the data will address this.
* Dataset doesn't include some vital stats such as Steal, Blocks or Turnovers. 


```python
min_max_scaler = preprocessing.MinMaxScaler()
games_scaled =  min_max_scaler.fit_transform(games[["PTS_home","FG_PCT_home","FT_PCT_home","FG3_PCT_home","AST_home","REB_home","TEAM_ID_away","PTS_away","FG_PCT_away","FT_PCT_away","FG3_PCT_away","AST_away","REB_away","PTS_total","AST_total","REB_total","HOME_TEAM_WINS"]])
games_scaled = pd.DataFrame(games_scaled)
games_scaled.columns = ["PTS_home","FG_PCT_home","FT_PCT_home","FG3_PCT_home","AST_home","REB_home","TEAM_ID_away","PTS_away","FG_PCT_away","FT_PCT_away","FG3_PCT_away","AST_away","REB_away","PTS_total","AST_total","REB_total","HOME_TEAM_WINS"]
games_scaled["HOME_TEAM_ID"] = games["HOME_TEAM_ID"]
games_scaled["VISITOR_TEAM_ID"] = games["VISITOR_TEAM_ID"]
games_scaled["SEASON"] = games["SEASON"]

print(games.shape, games_scaled.shape)

```

```python
for i in seasons:
    corr_year(i, games_scaled)
```

```python
games_scaled.corr().HOME_TEAM_WINS
```

When data is scaled to a [0,1] scale, the correlations can be more fairly analyzed. Here we see that the most correlated stat with winning is opponent FG%. Teams that win do so more often when the other team shoots poorly.

This in iteself is not surprising. Just about anyone could tell you that if your opponent shoots poorly that is good for you. It is interesting is that this correlation exceeds the benefits of a team shooting well themselves. It also exceeds the benefits of scoring more points. These four quantities, scoring and shooting percentage and their corresponding values for the opponent are the four most correalted values with winning games. 

These values exceed the correlations of Rebounds and Assists by a wide margin. While assists and rebounds, and preventing your opponents from getting them, are associated with winning this is substantially larger in magnitude than the correlation with points or FG%. 3P% is similarly significant, although less than FG%. FT% seems to have a lower level of importance.
